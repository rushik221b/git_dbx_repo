```python
# creating sample data frame for employee details
# importing required libraries :
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DateType
from datetime import date\

#creating schema : EmpId, Empname, Age, DeptId, Salary, JoinDate

schema = StructType([
                StructField("EmpId", IntegerType(), False),
                StructField("Empname", StringType(), True),
                StructField("Age", IntegerType(), True),
                StructField("Deptid", StringType(), True),
                StructField("Salary", IntegerType(), True),
                StructField("JoinDate", DateType(), True)
])

# Sample data :

data = [
    (1, "Amit Sharma", 28, "IT", 55000.00, date(2022, 3, 15)),
    (2, "Priya Singh", 32, "HR", 48000.00, date(2021, 7, 10)),
    (3, "Rahul Verma", 26, "Finance", 52000.00, date(2023, 1, 5)),
    (4, "Neha Gupta", 30, "Marketing", 60000.00, date(2020, 11, 20)),
    (5, "Vikram Patel", 35, "IT", 75000.00, date(2019, 6, 25))
]

# Creating dataFrame :

df = spark.createDataFrame(data, schema)

df.show()
```
